# 常用数据结构
pd.Series (序列) 可以看成是 DataFrame 的一个列
    value：返回 Series 对象所有元素
    index：返回索引
    dtypes：返回数据类型
    shape：返回 series 数据形状
    ndim：返回对象维度
    size：返回对象个数

pd.DataFrame (表格)  
    values：返回 Series 对象所有元素
    index：返回索引
    dtypes：返回每一列的数据类型
    shape：返回 series 数形状
    ndim：返回对象维度
    size：返回对象个数
    columns：返回列标签

````python
import numpy as np
import pandas as pd
data = np.random.randint(100, size=20)
print(data)

result = pd.cut(data, bins=[0,30,80,100], right=True)
print(result)

df = pd.DataFrame(data, columns=['data'])
df['data_type'] = result
df
````
